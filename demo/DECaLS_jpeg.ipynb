```python
# Import packages
%matplotlib inline
import os
import urllib

import numpy as np

import matplotlib.pyplot as plt

import astropy.wcs as wcs
from astropy.io import fits
from astropy.table import Table, Column
```

```python
def gen_url_decals(ra_cen, dec_cen, pix_scale, layer, size, bands):
    return 'http://legacysurvey.org/viewer/jpeg-cutout?ra=' + str(
        ra_cen) + '&dec=' + str(dec_cen) + '&layer=' + layer + '&size=' + str(
            size) + '&pixscale=' + str(pix_scale) + '&bands=' + bands
```
